# NONPARAMETRIC REGRESSION ANALYSIS

Nonparametric regression analysis is a statistical method used to model the relationship between two variables when the functional form of the relationship is unknown or cannot be described by a specific mathematical equation. Unlike parametric regression models that make assumptions about the form of the relationship, nonparametric regression does not rely on specific assumptions and allows for more flexibility in capturing complex patterns.

In nonparametric regression, the relationship between the predictor variable (independent variable) and the response variable (dependent variable) is estimated using the observed data. It does not assume a specific functional form, such as a linear or polynomial relationship, but rather seeks to estimate the underlying relationship directly from the data.

```r
install.packages("mblm")
library(mblm)
```

#### 13.11.1
The following are the heart rates (HR: beats/minute) and oxygen consumption values ($VO_{2}$:cal/kg/24 h) for nine infants with chronic congestive heart failure:

Compute $ (\beta_{1}), ((\beta(\beta_{0})))$ and $((\beta_{0}))_{2,M}$

```r
hr <- c(163, 164, 156, 151, 152, 167, 165, 153, 155)
vo <- c(53.9, 57.4, 41.0, 40.0, 42.0, 64.4, 59.1, 49.9, 43.2)


modelos <- mblm(formula= vo~hr, repeated = F)
regresion <- lm(vo~hr)
plot(vo~hr)
abline(modelos, col = "blue")
abline(regresion, col = "red")
summary(modelos)
```

Based on the nonparametric regression analysis using the median-based linear model (mblm), the following conclusions can be drawn:

1. The estimated intercept ($\beta_0$) is -176.6536 with a median absolute deviation (MAD) of 2.3898. The intercept represents the expected value of oxygen consumption when the heart rate is zero. The p-value for the intercept is 0.00391, indicating that it is statistically significant at the 0.05 significance level.

2. The estimated slope ($\beta_1$) is 1.4288 with a MAD of 0.9266. The slope represents the expected change in oxygen consumption per unit increase in heart rate. The p-value for the slope is 7.19e-05, indicating that it is highly statistically significant.

3. The $R^2$ value (V value) is 572, which indicates that 57.2% of the variability in oxygen consumption can be explained by the linear relationship with heart rate.

4. The residual standard error is 3.923, which provides an estimate of the variability of the observed oxygen consumption values around the regression line.

In conclusion, there is a significant linear relationship between heart rate and oxygen consumption in infants with chronic congestive heart failure. As the heart rate increases, the oxygen consumption tends to increase, as indicated by the positive slope coefficient. The model provides a reasonable fit to the data, explaining a substantial portion of the variability in oxygen consumption.


#### 13.11.2
The following are the body weights (grams) and total surface area (cm2) of nine laboratory
animals:

Compute the slope estimator and two intercept estimators.

```r
weight <- c(660.2, 706.0, 924.0, 936.0, 992.1, 888.9, 999.4, 890.3, 841.2)
area <- c(781.7, 888.7, 1038.1, 1040.0, 1120.0, 1071.5, 1134.5, 965.3, 925.0)


modelos <- mblm(formula= area~weight, repeated = F)
regresion <- lm(area~weight)
plot(area~weight)
abline(modelos, col = "blue")
abline(regresion, col = "red")
summary(modelos)
```

Based on the nonparametric regression analysis using the median-based linear model (mblm), the following conclusions can be drawn:

1. The slope estimator is 1.0094. This means that, on average, for every 1 gram increase in body weight, there is an estimated increase of 1.0094 cm² in total surface area. The p-value for the slope is 2.3e-06, indicating that it is highly statistically significant.

2. The intercept estimator can be calculated in two ways: The median-based estimator is 115.3152, and the mean-based estimator is not provided in the analysis.

3. The $R^2$ value (V value) is 608, which indicates that 60.8% of the variability in total surface area can be explained by the linear relationship with body weight.

4. The residual standard error is 40.9, which provides an estimate of the variability of the observed total surface area values around the regression line.

In conclusion, there is a significant linear relationship between body weight and total surface area in laboratory animals. As the body weight increases, the total surface area tends to increase, as indicated by the positive slope coefficient. The model provides a reasonable fit to the data, explaining a substantial portion of the variability in total surface area. However, the mean-based intercept estimator is not provided, limiting our understanding of the intercept.
